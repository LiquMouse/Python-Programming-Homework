```python
#代码
X = np.random.randint(0,100,size=100)
sinX = np.sin(X)
print(X)
print(sinX)
```

### 1.3 随机生成100个范围在0到100之间的整数，统计大于40小于60的元素个数

```python
#代码
X = np.random.randint(0,100,size=100)
Y = np.bincount(X)
Z = np.sum(Y[41:60])
print(Y)
print(Z)
```
