## 3、matplotlib基础（选做）

```python
#数据准备
x=np.linspace(0,8*np.pi,100) 
y_s=np.sin(x)
y_c=np.cos(x)
y_r=(np.random.random(100)-0.5)*2 #随机数
```

### 3.1 根据下图和上面给出的数据，写出绘制代码

```python
#开始绘图
fig = plt.figure(figsize=(15,5))
#添加子图
axes = fig.add_subplot(1,1,1)
#设置坐标系的标题
axes.set_title("这是一个demo",fontsize=25)

#画图
axes.plot(x, y_s, 'r--', label='sin(x)')
axes.plot(x, y_c, 'b--', label='cos(x)')
axes.plot([0,25], [1,1], '-.')
axes.plot(x, y_r, color='grey', linestyle='', marker='1')

#设置y轴范围
plt.ylim(-2,2)

#加图例
axes.legend(loc='upper right')
#加注释
axes.annotate('这是指向性注释',(12.5,1),(13,1.5),arrowprops={},fontsize=14,color='r')
axes.text(15,-1.5,"这是无指向注释",fontsize=14)
#xy轴
plt.xlabel('x轴',color='b',fontsize=18)
plt.ylabel('y轴',color='r',fontsize=18)
#网格
axes.grid(linestyle='--')
#参考区域
axes.axvspan(3.3,4.5,0.2,0.8)
```
