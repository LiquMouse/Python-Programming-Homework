### 练习1：异常处理——运算异常
> 实现一个除法计算器，提示用户输入除数和被除数，然后把两个数的相除的结果进行输出，要求只有当用户输入quit的时候系统才会退出，否则一直可以进行除法运算。当用户输入非数字（quit除外）或者除数为0时，捕获相应的异常，并将异常信息写入日志文件

```python
# 代码
import time

while True:
    f = open("log.txt","a")
    a = input("输入除数或输入quit退出: ")
    if a == 'quit':
        break
    b = input("输入被除数或输入quit退出: ")
    if b == "quit":
        break
    try:
        b = float(b)
        a = float(a)
        print(b/a)
    except ZeroDivisionError as t:
        print("除数不能为零!")
        now = (time.strftime("%Y-%m-%d %H:%M:%S", time.localtime()))
        f.write(f"{now}: {t}")
        continue
    except ValueError as t:
        print("请输入数字或quit!")
        now = (time.strftime("%Y-%m-%d %H:%M:%S", time.localtime()))
        f.write(f"{now}: {t}")
        continue
    except Exception as t:
        print(f"异常: {t}")
        now = (time.strftime("%Y-%m-%d %H:%M:%S", time.localtime()))
        f.write(f"{now}: {t}")
        continue
    

```

### 练习2：异常处理——文件异常
> 改写上机5中的练习1，添加相应的异常处理代码，确保程序不会因为异常而中断



```python
# 代码
import os

def find_files(path, *last, frag=""): # path: 路径; last: 后缀, 可变参数; frag: 文件名片段, 只能用关键字传递
    try:
        # 搜索目录和子目录中所有文件
        # 之所以写个嵌套函数, 是因为last为可变参数, 如果直接调用自身, 那么会给last再加一层括号形成嵌套元组, 
        # 比如last=('py','txt)会变成last=(('py','txt')), 导致程序出错. 于是定义一个没有可变参数的函数
        def find_all_files(path=path, lasts=last, frag=frag): 
            ls = [] # 用来存符合要求的文件路径

            # 遍历指定目录下的文件/文件夹
            for file in os.listdir(path):
                full_path = os.path.join(path,file) # 文件完整路径
                # 如果为文件, 且符合要求, 则将路径加入ls
                if os.path.isfile(full_path):
                    for i in last:
                        if frag in file and file.endswith(i):
                            ls.append(full_path)
                
                # 如果为文件夹, 则在该子目录下搜索, 搜索结果与当前ls合并
                else:
                    ls += find_all_files(full_path, lasts, frag=frag)

            return ls
        
        return find_all_files()
    except Exception as t:
        return t

files = find_files("D:\\Users\\Wuhlulu\\Document", ".py", ".txt", frag='1')
print(files)
```

### 练习3：面向对象编程——三角形
> 定义一个三角形类，通过传入三条边来构造三角形，并提供计算周长和面积的方法

```python
#代码
class triangle():
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
    def perimeter(self): # 算周长
        return self.a + self.b + self.c
    def area(self): # 算面积
        p = (self.a + self.b + self.c) / 2
        s = (p*(p-self.a)*(p-self.b)*(p-self.c))**0.5
        return s
    
tri = triangle(4,5,6)
print(tri.perimeter())
print(tri.area())
```
